```python
records = sf.records()
pits_inds = []
for i in range(len(records)):
    record = records[i]
    #if int(record[0]) in char_zipcodes:   #for zipcodes
    idnum = int(record[0][-11:])
    if idnum in char_cts:
        pits_inds.append(i)
```

```python
pits_shapeRecs = []
for ind in pits_inds:
    shapeRec = sf.shapeRecord(ind)
    pits_shapeRecs.append(shapeRec)
```

```python
w = shapefile.Writer(shapeType=sf.shapeType)
w.fields = sf.fields
for shapeRec in pits_shapeRecs:
    w.shapes().append(shapeRec.shape)
    w.records.append(shapeRec.record)
```

```python
w.save('char_ct_shapefile')
```
